```python
from transformers import AutoTokenizer , AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("camembert-base")
model = AutoModelForMaskedLM.from_pretrained("camembert-base")
```

```python
en = "Je prépare le dîner tous les <mask> :)"

seq = tokenizer(en , return_tensors = "pt")
```

```python
seq
```

```python
import torch
```

```python
with torch.no_grad():
    output = model(**seq)
```

```python
mask_token = (seq.input_ids == tokenizer.mask_token_id).nonzero(as_tuple = True)[1]
mask_token_logits = output.logits[0,mask_token,:]
predicted_token = mask_token_logits.argmax(axis=-1)

word = tokenizer.decode(predicted_token)


```
