```python
from transformers import AutoTokenizer 

```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

se = "i love this course by huggingface"

tokens = tokenizer.tokenize(se)

token = tokenizer.convert_tokens_to_ids(tokens)
print(tokens)
print(token)
```
