```python
from transformers import AutoTokenizer , AutoModelForSequenceClassification

import torch
```

```python
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)

model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python
seq1 = "I have been waiting for this hugging face course my whole life"

seq2 = "I hate this so much"
```

```python
tokens1 = tokenizer.tokenize(seq1)

tokens2 = tokenizer.tokenize(seq2)
```

```python
id1 = tokenizer.convert_tokens_to_ids(tokens1)
id2 = tokenizer.convert_tokens_to_ids(tokens2)
```

```python
inputid1 = torch.tensor([id1])
inputid2 = torch.tensor([id2])
```

```python
print(inputid1,inputid2)
```

```python
output1 = model(inputid1)

output2 = model(inputid2)

print(output1.logits , output2.logits)
```

```python
sequences = ["I have a dog that is cuter than a cat", " I have a pet camel and his name is harry" , "I have a pet snake"]

tokens = tokenizer(sequences , truncation = True , padding = True , return_tensors = "pt")
```

```python
output = model(**tokens)
```
